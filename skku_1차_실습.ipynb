```python
import pandas as pd

# CSV 파일 불러오기
data = pd.read_csv("/content/drive/MyDrive/황수웅/sci/SVM/SVM.csv")
data
```

```python
data = data.drop(data.columns[0], axis=1)
print(data.info())
```

```python
import matplotlib.pyplot as plt
# 특정 변수를 활용하여 plot 그리기
# 변수를 활용하여 산점도 그리기
plt.scatter(data['Height'], data['Weight'])
plt.xlabel('Height')  # X 축 라벨 설정
plt.ylabel('Weight')  # Y 축 라벨 설정
plt.title('Scatter Plot of H and W')  # 그래프 제목 설정
plt.show()
```

```python
# 두 변수의 상관계수 계산
correlation = data[['Height', 'Weight']].corr()
print(correlation)
```

```python
# BMI, Height, Weight
# 필요한 라이브러리 가져오기
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# 독립변수와 종속변수 분리
X = data[['Height', 'Weight']]
y = data['BMI']

# 데이터를 학습용과 테스트용으로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 데이터 표준화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVR 모델 훈련
svr_model_rbf = SVR(kernel='rbf')
svr_model_rbf.fit(X_train_scaled, y_train)

# SVR 모델 훈련
svr_model_lin = SVR(kernel='linear')
svr_model_lin.fit(X_train_scaled, y_train)

# 각 모델의 예측
y_pred_svr_rbf = svr_model_rbf.predict(X_test_scaled)
y_pred_svr_lin = svr_model_lin.predict(X_test_scaled)

# 각 모델의 성능 비교 (MSE 사용)
mse_svr_rbf = mean_squared_error(y_test, y_pred_svr_rbf)
mse_svr_lin = mean_squared_error(y_test, y_pred_svr_lin)

print("MSE of SVR_rbf Model:", mse_svr_rbf)
print("MSE of SVR_linear Model:", mse_svr_lin)
```

```python
import matplotlib.pyplot as plt

# SVR_rbf 모델 예측 결과 시각화
plt.figure(figsize=(5, 3))
plt.plot(y_test, y_pred_svr_rbf, 'o', color='red', label='SVR_rbf Model', markersize=8)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)
plt.xlabel('Actual BMI')
plt.ylabel('Predicted BMI')
plt.title('SVR_rbf Model Prediction')
plt.legend()
plt.grid(True)
plt.show()

# SVR_linear 모델 예측 결과 시각화
plt.figure(figsize=(5, 3))
plt.plot(y_test, y_pred_svr_lin, 'o', color='orange', label='SVR_linear Model', markersize=8)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)
plt.xlabel('Actual BMI')
plt.ylabel('Predicted BMI')
plt.title('SVR_linear Model Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
